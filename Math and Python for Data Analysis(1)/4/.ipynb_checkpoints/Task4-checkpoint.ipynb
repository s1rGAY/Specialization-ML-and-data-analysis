```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Генерация выборки

Гистограмма выборки и теоретическая оценка плотности вероятности распределения Лапласа

```python
lapl = sts.laplace(0,1)
rvs = lapl.rvs(1000)

x = np.linspace(-7.5,7.5,1000)
# значение функции распределения Лапласа:
cdf_lapl = lapl.cdf(x)
# значение функции плотности вероятности Лапласа:
pdf_lapl = lapl.pdf(x)

plt.plot(x, pdf_lapl, label='Laplace',linewidth = 2.5)
plt.hist(rvs, density = True)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```
