# Распределение Вейбулла

Распределение Ве́йбулла в теории вероятностей — двухпараметрическое семейство абсолютно непрерывных распределений. Названо в честь Валодди Вейбулла, детально охарактеризовавшего его в 1951, хотя впервые его определил Фреше в 1927, а применено оно было ещё в 1933 для описания распределения размеров частиц.

Если интересно, вот видео по закону распределения Вейбулла - https://www.youtube.com/watch?v=Dprt6L7c0Hc

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Задание распределения
k, lamb = 1, 1.5
weib_rv = sts.exponweib(k, lamb)

#выборка объема 1000 из Вейбулловского распределения
sample = weib_rv.rvs(1000)

#При заданных k и lamb лучше проводить генерацию в пределах от 0 до 5
x = np.linspace(0,5,1000)


pdf_weib = weib_rv.pdf    #Ф-ия плотности распределения для Вейбулла
pdf_weib = weib_rv.pdf(x) #значение ф-ии плотности вероятности
```

```python
# построим гистограмму выборки и наложим на нее теоретическую 
# оценку плотности вероятности экспоненциального распределения

# density=True представляет собой распределение плотности вероятности
# density=False представляет собой частотное распределение

# рисуем гистограмму распределения и функцию плотности:
plt.plot(x, pdf_weib, label='Weibull',linewidth = 2.5)
plt.hist(sample, density = True)

# ДИСПЕРСИЯ
var = (sts.exponweib(k, lamb)).var()

#математическое ожидание данного распределения
mean = weib_rv.mean()

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='lower right')

print(f'Распределение Вейбулла с параметрами K = {k} и lambda = {lamb}. \
Матожидание {round(mean, 5)} Дисперсия {round(var, 5)}')
```

Как видно на графике выше, гистограмма практически совпадает с теор. оценкой плотности вероятности.

# Апроксимация

```python
# ф-ия для генерации выборок объема n
def mk_pdf(n): 
    #средние по выборкам объема n
    means = []
    for _ in range(1000):
        sample = weib_rv.rvs(n)
        sample_means = np.mean(sample)
        means.append(sample_means)
    
    #pdf для распределения Вейбула
    # mu = mean!
    mu,sigma = mean, np.sqrt(var / n)
    
    #значение ф-ии нормального распределения
    pdf = sts.norm(mu, sigma).pdf
    
    # построение гистограммы и теор. оценки плотности вероятности нормального распределения на ней
    x = np.linspace(0,2,1000)
    
    plt.plot(x, pdf(x), linewidth = 1, color = 'r')
    plt.hist(means, density=True)
    plt.xlabel('$\\bar{X}_n$')
    plt.ylabel('$f(\\bar{X}_n)$')
    plt.title('n = %d' % n)
    plt.show()
```

```python
#построение графиков при помощи ранее написанной ф-ии generate_pdf
for n in [5, 10, 50, 100, 500, 1000, 10000]:
    mk_pdf(n)
```

### Вывод
Заметно, что при увеличении размера выборки гистограмма начинает точнее напоминать нормальное распределение, а также можно увидеть, что дисперсия понижается (видно из графиков). 

Видно, что Центральная предельная теорема работает при больших n (что и следовало ожидать).
